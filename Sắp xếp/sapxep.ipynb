```python
# Đọc lại file dataset sau khi môi trường đã được reset
import pandas as pd

# Đọc lại file dataset
df = pd.read_csv('covereal1.csv')

# Lấy tất cả cột nguyên liệu (loại bỏ cột 'Dish Name')
ingredient_columns = [col for col in df.columns if col != 'Dish Name']

# Đếm số lần xuất hiện của mỗi nguyên liệu
ingredient_counts = df[ingredient_columns].sum(axis=0).reset_index()
ingredient_counts.columns = ['Ingredient', 'Count']

# Lưu kết quả vào một file CSV mới
ingredient_counts.to_csv('ingredient_counts.csv', index=False)

# Hiển thị kết quả
ingredient_counts.head()
```
