chỉ nên để 15% là từ 1 đến 2s thôi, quá thì bị lmao wa, còn lại random 3 đến 5s

```python
import pandas as pd
import numpy as np

df = pd.read_csv("dataset/final_dataset.csv")

# Xác suất xuất hiện của từng mức rating
ratings = [1, 2, 3, 4, 5]
probabilities = [0.075, 0.075, 0.3, 0.3, 0.25]  # 15% cho rating từ 1 đến 2, còn lại từ 3 đến 5

df["rating"] = np.random.choice(ratings, size=len(df), p=probabilities)

# Lưu tập dữ liệu mới
df.to_csv("final_dataset_with_ratings.csv", index=False)

print("New column 'rating' filled with weighted random values!")
```
