Fat (Cht bo): Nu > 20g, cn hn ch.

Saturates (Cht bo bo ha): Nu > 5g, cn hn ch.

Carbs (Carbohydrate): Nu > 50g, cn hn ch.

Quy tc to ct "Hypertension":

Ct "Hypertension = 0": Nu c bt k nguyn liu no trong nhm cn hn ch hoc nu ngng ca cc ct Salt, Fat, Saturates, hoc Carbs vt qua cc gi tr hn ch hoc khng c cc nguyn liu u tin

Ct "Hypertension = 1": Nu khng c nguyn liu no trong nhm cn hn ch v cc ngng mui, cht bo, cht bo bo ha, carbs u khng vt qua hn ch, nhng c t nht mt nguyn liu trong nhm u tin.

Quy trnh thc hin:

Bc 1: Kim tra cc nguyn liu cn hn ch trong mn n.

Bc 2: Kim tra cc ngng dinh dng (mui, cht bo, cht bo bo ha, carbohydrate).

Bc 3: Da trn cc iu kin trn, gn gi tr cho ct "Hypertension".

```python
# Kim tra s tn ti ca cc nguyn liu t 2 list (preferred_ingredients v restricted_ingredients)
existing_ingredients = [ingredient for ingredient in preferred_ingredients + restricted_ingredients if ingredient in df.columns]

# Tm cc nguyn liu khng c trong dataset
missing_ingredients = [ingredient for ingredient in preferred_ingredients + restricted_ingredients if ingredient not in df.columns]

# Nu khng tm thy nguyn liu no t danh sch, in ra cc nguyn liu thiu
if missing_ingredients:
    print("Cc nguyn liu khng tn ti trong dataset:")
    print(missing_ingredients)
else:
    print("Tt c cc nguyn liu tn ti trong dataset.")

```

```python
# nh ngha cc ngng cho phn loi
high_salt_threshold = 1.5
high_fat_threshold = 20
high_saturates_threshold = 5
high_carbs_threshold = 50

# Tnh ton Hypertension theo ng quy tc
df['Hypertension'] = (
    (df['Salt'] <= high_salt_threshold) &
    (df['Fat'] <= high_fat_threshold) &
    (df['Saturates'] <= high_saturates_threshold) &
    (df['Carbs'] <= high_carbs_threshold) &
    (df[restricted_ingredients].any(axis=1) == False) &
    (df[preferred_ingredients].any(axis=1) == True)
).astype(int)

# Hin th kt qu u tin
print(df[['Dish Name', 'Hypertension']].head())
```

```python
def classify_heart_healthy(row):
    """
    Phn loi mn n c tt cho tim mch hay khng
    Heart-Healthy Diet = 1 (Tt), = 0 (Khng tt)
    """
    
    # Kim tra cc tiu ch dinh dng
    fat_ratio = (row['Fat'] / row['Calories']) * 100
    saturates_ratio = (row['Saturates'] / row['Calories']) * 100
    sugars_ratio = (row['Sugars'] / row['Calories']) * 100
    salt = row['Salt']
    fibre = row['Fibre']
    
    # iu kin dinh dng cho Heart-Healthy
    nutrition_good = (fat_ratio <= 35 and 
                     saturates_ratio <= 10 and 
                     sugars_ratio <= 10 and 
                     salt <= 1 and 
                     fibre >= 2.5)
    
    heart_healthy_ingredients = [
   # C giu Omega-3
   'salmon', 'mackerel', 'sardines', 'tuna', 'anchovy fillets', 'smoked trout',
   
   # Rau xanh m mu
   'spinach', 'kale', 'broccoli', 'chard', 'bok choi', 'watercress', 'rocket',
   'endive', 'chicory', 'cress',
   
   # Tri cy h berry
   'blueberries', 'strawberries', 'raspberries', 'blackberries', 'cranberries', 'berries',
   
   # Tri cy tt khc
   'cherries', 'pomegranate', 'grapes', 'orange', 'grapefruit', 'apricot', 
   'peach', 'nectarine', 'plum', 'kiwi', 'melon', 'watermelon',
   
   # Cc loi ht
   'almonds', 'walnuts', 'pistachios', 'pecans', 'cashews', 'hazelnuts',
   
   # Ng cc nguyn ht
   'oats', 'quinoa', 'pearl barley',
   
   # Du olive
   'olive oil',
   
   # u v cc loi u
   'lentil', 'chickpeas', 'broad beans', 'borlotti beans', 'green beans', 'bean',
   
   # Cc nguyn liu tt khc
   'avocado', 'garlic', 'ginger', 'turmeric', 'beetroot', 'beets', 'carrot',
   'asparagus', 'pepper', 'linseed', 'chia', 'pumpkin seeds', 'sesame seeds',
   'pumpkin', 'sesame',
   
   # Tho mc/gia v tt
   'basil', 'oregano', 'thyme', 'rosemary', 'parsley', 'mint', 'coriander',
   'cumin', 'cinnamon'
]

    heart_restricted_ingredients = [
    # Tht  v tht ch bin
    'bacon', 'beef', 'beef mince', 'ham', 'lamb', 'lamb mince', 'lamb cutlets',
    'lamb shoulder', 'pork', 'pork mince', 'pork ribs', 'pork shoulder', 'sausage',
   'prosciutto', 'pancetta', 'chorizo', 'bresaola', 'mortadella', 'pastrami',
   'salami', 'nduja', 'speck',
   
   # Cht bo bo ha
   'butter', 'cream', 'buttercream', 'buttermilk', 'lard',
   
   #  ngt
   'sugar', 'brown sugar', 'caster sugar', 'demerara sugar', 'icing sugar',
   'golden syrup', 'syrup', 'treacle', 'jam', 'honey', 'chocolate', 
   
   # Thc phm mn/ch bin
    'soy sauce', 'fish sauce', 'pickle',
   
   # Ph mai nhiu sodium
   'cheddar', 'parmesan', 'stilton', 'camembert',
   
   #  ung c cn
   'beer', 'wine', 'whisky', 'gin', 'vodka'
]
    
    # Kim tra c nguyn liu u tin
    has_priority = any(row.get(ing, 0) == 1 for ing in heart_healthy_ingredients)
    
    # Kim tra c nguyn liu hn ch
    has_restricted = any(row.get(ing, 0) == 1 for ing in heart_restricted_ingredients)

    # Phn loi
    if nutrition_good and has_priority and not has_restricted:
        return 1  # Heart-Healthy
    else:
        return 0  # Khng Heart-Healthy

# To ct Heart-Healthy Diet s dng hm classify_heart_healthy
df['Heart-Healthy Diet'] = df.apply(classify_heart_healthy, axis=1)
```
